#<u>Understanding Digital Images for python processing</u>

```python
#we use the library scikit-image and use the function io from library
import skimage 
from skimage import io
```

On printing the image, we get the NumPy array of the image.

The below printed NumPy array is of the form [R G B][x y z]
where x, y, z are co-ordinates and R-red, G-green, B-blue are corresponding values of the color.

```python
my_image = io.imread("/content/covid.jpg")
print(my_image)
```

```python
#to see the image
plt.imshow(my_image)
```

```python
#we see the image is of 1600 by 2048 pixels and 3 stands for RGB
print(my_image.shape)
#the datatype of the entries to the numpy ndarray
#uint means unsigned integer and each entry is of 8 bits
#i.e., can go from 0 to 255
print(my_image.dtype)
```

```python
#min and max are w.r.t. uint8 i.e., unsigned 8 bit integer
#min of uint8 is 0 and max is 2^8-1= 256-1=255
print(my_image.min(), my_image.max())
```

Now, let's create an artificial image 

```python
import numpy as np
```

```python
#to show images we will use matplotlib's pyplot 
from matplotlib import pyplot as plt
random_image = np.random.random([500,500]) #the dimensions are 500x500
plt.imshow(random_image)
print(random_image)
```

```python
print(random_image.shape)
#image is of the 500 by 500 pixel
print(random_image.dtype)
#now the entry is ndarray is of type float and each entry 
#has 64 bits.
```

```python
#min() prints the minimum value of a numpy array
#min() and max() are w.r.t. float64
print(random_image.min(), random_image.max())
```

We can convert the type of the image. We can convert an int type image to a floating point image.

```python
from skimage import img_as_float
```

```python
#converting my_image from int to float typt image
my_float_image = img_as_float(my_image)
```

```python
#the min and max for the int converted to float image is between 0.0 and 1.0
print(my_float_image)
print(my_float_image.min(), my_float_image.max())
plt.imshow(my_float_image)
```

Here, we notice that the int image is same as the float image but the min and max of int image is from 0 to 255 and that of float image is from 0.0 to 1.0

When changing the data type from let's say 64 bits to 16 bits we will be losing on some information present in the pixel if information is stored within those pixels, called as down-scaling an image.

dtypes...

uint8 = 0 to 255  
uint16 = 0 to 65535  
uint32 = 0 to ((2**32) - 1)  
float = -1 to 1 or 0 to 1  
int8 = -128 to 127  
int16 = -32768 to 32767  
int32 = -2**31 to 2**31 - 1  

Functions that convert images to desired dtype and properlyrescale their values  
img_as_float - Convert to 64-bit floating point.  
img_as_ubyte - Convert to 8-bit uint.  
img_as_uint - Convert to 16-bit uint.  
img_as_int - Convert to 16-bit int.
